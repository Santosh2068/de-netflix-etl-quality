```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned/netflix_cleaned.csv")
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False).head(15)
```

```python
df["type"].value_counts().plot(kind="bar", title="Count: Movies vs TV Shows")
plt.show()
```

```python
df["release_year"].plot(kind="hist", bins=30, title="Release year distribution")
plt.show()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
by_year = df.dropna(subset=["date_added"]).groupby(df["date_added"].dt.year)["show_id"].count()
by_year.plot(kind="line", title="Titles added by year (based on date_added)")
plt.show()
```
